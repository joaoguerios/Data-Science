<h1>Polynomial Linear Regression</h1>
<p>y = b0 + b1*x1 + b2*x**2 + b3*x**3 + bn*x**n 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder # Scaling categorical
from sklearn.model_selection import train_test_split #spliting
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("Data.csv")
df_infos = df.iloc[:,1:-1]
df_results = df.iloc[:,-1]
```

linear regression

```python
linear = LinearRegression()
linear.fit(df_infos,df_results)
df_linear_pred = linear.predict(df_infos)
```

```python
plt.scatter(df_infos,df_results, color = "red", label = "Data")
plt.plot(df_infos, df_linear_pred, color = "blue", label = "results")
plt.title("linear Regression")
plt.xlabel("level")
plt.ylabel("Salary")
plt.show()
```

polynomial regression

```python
polynomial = PolynomialFeatures(degree=2)
df_infos_poly = polynomial.fit_transform(df_infos)
polynomial_reg = LinearRegression()
polynomial_reg.fit(df_infos_poly,df_results)
df_polynomial_pred = polynomial_reg.predict(df_infos_poly)
```

```python
plt.scatter(df_infos,df_results, color = "red", label = "Data")
plt.plot(df_infos, df_polynomial_pred, color = "blue", label = "results")
plt.title("Polynomial")
plt.xlabel("level")
plt.ylabel("Salary")
plt.show()
```

Improving the results from polynomial

```python
polynomial_v2 = PolynomialFeatures(degree=4)
df_infos_poly = polynomial_v2.fit_transform(df_infos)
polynomial_reg_v2 = LinearRegression()
polynomial_reg_v2.fit(df_infos_poly,df_results)
df_polynomial_pred_v2 = polynomial_reg_v2.predict(df_infos_poly)
```

```python
plt.scatter(df_infos,df_results, color = "red", label = "Data")
plt.plot(df_infos, df_polynomial_pred_v2, color = "blue", label = "results")
plt.title("Polynomial")
plt.xlabel("level")
plt.ylabel("Salary")
plt.show()
```

Predicting a salary with the polynomial_v2 

```python
polynomial_reg_v2.predict(polynomial_v2.fit_transform([[6.5]]))
```
