<h2>Libraries</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split # split in train and test
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

<h2>DataSet</h2>

```python
df = pd.read_csv("Data.csv")
df_infos = df.iloc[:,:-1].values
df_results = df.iloc[:,-1].values
print(df_infos)
print(df_results)
```

<h3>Spliting the Dataset</h3>

```python
df_info_train,df_info_test,df_results_train,df_results_test = train_test_split(df_infos,df_results,test_size= 0.3, random_state= 1)
```

<h2>Training</h2>

```python
regressor = LinearRegression()
regressor.fit(df_info_train,df_results_train)
```

<h2>Predicting</h2>

```python
df_train_pred = regressor.predict(df_info_train)
df_test_pred = regressor.predict(df_info_test)
```

<h4>Training plot</h4>

```python
plt.scatter(df_info_train,df_results_train, color = "red", label = "Data")
plt.plot(df_info_train, df_train_pred, color = "blue", label = "results")
plt.title("Training Set")
plt.xlabel("Years of Exp")
plt.ylabel("Salary")
```

<h4>Test plot</h4>

```python
plt.scatter(df_info_test,df_results_test, color = "red", label = "Data")
plt.plot(df_info_test, df_test_pred, color = "blue", label = "results")
plt.title("Test Set")
plt.xlabel("Years of Exp")
plt.ylabel("Salary")
```

<h2>Model evaluation</h2>

```python
mse_train = mean_squared_error(df_results_train, df_train_pred)
r2_train = r2_score( df_results_train, df_train_pred)
mse_test = mean_squared_error(df_results_test,df_test_pred )
r2_test = r2_score(df_results_test, df_test_pred)

print(f'MSE Treino: {mse_train:.2f}')
print(f'R2 Treino: {r2_train:.2f}')
print(f'MSE Teste: {mse_test:.2f}')
print(f'R2 Teste: {r2_test:.2f}')
```
